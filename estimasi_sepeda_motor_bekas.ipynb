# deskripsi library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Memanggil Dataset

```python
df_raw = pd.read_csv('BIKE_DETAILS.csv')
```

# menghapus data yang bernilai empty pada Dataset

```python
df = df_raw.dropna()
```

# Deskripsi Dataset

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

```python
print(df.columns)
```

# Jumlah motor berdasarkan tahun

```python
motor = df.groupby('name').count()[['year']].sort_values(by='year',ascending=True).reset_index()
motor = motor.rename(columns={'year':'numberOfMotors'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=motor['name'], y=motor['numberOfMotors'], color='royalblue')
plt.xticks(rotation=60)
```

# Distribusi Jumlah kilometer yang telah ditempuh sepeda motor

```python
plt.figure(figsize=(15,5))
sns.distplot(df['km_driven'])
```

# Distribusi Harga showroom sepeda motor


```python
plt.figure(figsize=(15,5))
sns.distplot(df['ex_showroom_price'])
```

# Distribusi Harga jual motor

```python
plt.figure(figsize=(15,5))
sns.distplot(df['selling_price'])
```

# seleksi fitur

```python
features = ['year','km_driven','ex_showroom_price']
x = df[features]
y = df['selling_price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#year=2011, km_driven=35000, ex_showroom_price=70433
input_data = np.array([[2011, 35000, 70433]])

prediction = lr.predict(input_data)
print('Estimasi harga motor dalam US :', prediction)
```

# save model
